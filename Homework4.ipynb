First we need to make a large number of boostrapped samples from the original sample. After that, calculate the mean for each bootstrapped sample, and we can have a distribution of bootstrapped means. Then, using the python function np.quantile(bootstrapped_means, 0.025) and np.quantile(bootstrapped_means, 0.975) to get 95% confidence intervel of the bootstrapped means. By this way, we don't need to calculate the standard error of the mean. It is noticeable that we cant change size for each boostrapped sample because for example increase the sample size will lead to a decrease on the variability in that sample mean for samples of that size. Besides, 'with replacement' is important otherwise we will get a same sample mean for every bootstrapped mean.

```python
import numpy as np

sample = np.array([4, 8, 15, 16, 23, 42, 50, 7, 12, 22])  #sample

num_bootstrap_samples = 1000 #bootstrap number

bootstrapped_means = [] #store bootstrap mean

# Perform the bootstrapping process
for _ in range(num_bootstrap_samples):
    # Step 1: Resample from the original data WITH replacement
    bootstrapped_sample = np.random.choice(sample, size=len(sample), replace=True)
    
    # Step 2: Calculate the statistic (mean in this case) for each bootstrapped sample
    bootstrapped_mean = np.mean(bootstrapped_sample)
    
    # Step 3: Append the statistic (mean) to the list
    bootstrapped_means.append(bootstrapped_mean)

# Step 4: Calculate the 95% confidence interval (using the 2.5th and 97.5th percentiles)
lower_bound = np.quantile(bootstrapped_means, 0.025)  # 2.5th percentile
upper_bound = np.quantile(bootstrapped_means, 0.975)  # 97.5th percentile

# Print the 95% confidence interval for the mean
print(f"95% Bootstrap Confidence Interval for the Mean: [{lower_bound}, {upper_bound}]")

```

```python
import numpy as np

sample = np.array([4, 8, 15, 16, 23, 42, 50, 7, 12, 22])  #sample

num_bootstrap_samples = 1000 #bootstrap number

bootstrapped_means = [] #store bootstrap mean

# Perform the bootstrapping process
for _ in range(num_bootstrap_samples):
    # Step 1: Resample from the original data WITH replacement
    bootstrapped_sample = np.random.choice(sample, size=len(sample), replace=True)
    
    # Step 2: Calculate the statistic (median in this case) for each bootstrapped sample
    bootstrapped_median = np.median(bootstrapped_sample) # changed to median
    
    # Step 3: Append the statistic (median) to the list
    bootstrapped_means.append(bootstrapped_median)# changed to median

# Step 4: Calculate the 95% confidence interval (using the 2.5th and 97.5th percentiles)
lower_bound = np.quantile(bootstrapped_means, 0.025)  # 2.5th percentile
upper_bound = np.quantile(bootstrapped_means, 0.975)  # 97.5th percentile

# Print the 95% confidence interval for the mean
print(f"95% Bootstrap Confidence Interval for the Median: [{lower_bound}, {upper_bound}]")

```

Difference between Standard Error of the Mean (SEM) and Standard Deviation (SD):

The standard deviation (SD) measures the spread of individual data points around the mean, while the standard error of the mean (SEM) estimates the variability of the sample mean as an estimator of the population mean.
The SEM is smaller than the SD, shrinking as the sample size increases, reflecting more confidence in the sample mean's approximation of the population mean.
Using SEM to Create a 95% Confidence Interval:

A confidence interval can be constructed as sample mean ± 2 times the SEM, which captures approximately 95% of bootstrapped sample means.
The variability of sample mean estimates depends on sample size. Larger samples result in lower SEM and narrower confidence intervals.
Constructing a 95% Bootstrapped Confidence Interval:

Bootstrapping is used to resample the data with replacement, creating many bootstrapped sample means.
The 95% confidence interval is created using the 2.5th and 97.5th percentiles of the bootstrapped means, covering exactly 95% of them, unlike the SEM approach, which is an approximation.
Python’s np.quantile() function can be used to calculate these percentiles, allowing a precise range for the population mean.
Why Bootstrapped Samples Must Match the Original Sample Size:

Bootstrapped samples need to be the same size as the original sample to accurately estimate the variability of the sample statistic.
If we sampled without replacement, the bootstrapped sample would have the same data points in a different order, leading to the same mean, providing no additional information about variability.
Bootstrapping Other Statistics (e.g., Median):

The same bootstrapping process applies to other statistics, such as the median.
To compute a confidence interval for the median, we replace np.mean() with np.median() in the bootstrapping loop.
What a 95% Confidence Interval for the Median Means:

A 95% confidence interval for the median gives the range within which the true population median is likely to lie with 95% confidence.
It reflects the uncertainty in estimating the population median based on a sample.
Distinguishing Between Population Parameters and Sample Statistics:

Population parameters (e.g., population mean, median) are fixed but unknown values describing the entire population.
Sample statistics (e.g., sample mean, sample median) are estimates derived from a sample, subject to variability.
Confidence intervals use sample statistics to infer population parameters, accounting for sampling variability. For example, the sample mean is used to estimate the population mean, and the confidence interval represents where the population mean is likely to fall.

```python
# Writing to a CSV file
import pandas as pd

data = """PatientID,Age,Gender,InitialHealthScore,FinalHealthScore
1,45,M,84,86
2,34,F,78,86
3,29,M,83,80
4,52,F,81,86
5,37,M,81,84
6,41,F,80,86
7,33,M,79,86
8,48,F,85,82
9,26,M,76,83
10,39,F,83,84"""

with open("patient_data.csv", "w") as file:
    file.write(data)
```

```python
# Read the CSV file
df = pd.read_csv("patient_data.csv")
df
```

```python
#create a column to show the change in each patient's score and this is used to show the function of the vaccine
df['HealthScoreChange'] = df['FinalHealthScore'] - df['InitialHealthScore']
df
```

Null Hypothesis: the vaccine has no effect on patients

```python
# Create a histogram of the HealthScoreChange

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.hist(df['HealthScoreChange'], bins=10, edgecolor='black')
plt.title('Histogram of Health Score Changes')
plt.xlabel('Change in Health Score')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
#turn this column to a list
change_list = df['HealthScoreChange'].tolist()
print(change_list)
```

```python
# Calculate the mean of HealthScoreChange
mean_health_score_change = df['HealthScoreChange'].mean()
print("Mean of Health Score Change:", mean_health_score_change)
```

```python
#bootstrap
import numpy as np

np.random.seed(520) #random seed set

sample = np.array(change_list)  #sample

num_bootstrap_samples = 1000 #bootstrap number

bootstrapped_means = [] #store bootstrap mean

# Perform the bootstrapping process
for _ in range(num_bootstrap_samples):
    # Step 1: Resample from the original data WITH replacement
    bootstrapped_sample = np.random.choice(sample, size=len(sample), replace=True)
    
    # Step 2: Calculate the statistic (mean in this case) for each bootstrapped sample
    bootstrapped_mean = np.mean(bootstrapped_sample)
    
    # Step 3: Append the statistic (mean) to the list
    bootstrapped_means.append(bootstrapped_mean)

# Step 4: Calculate the 95% confidence interval (using the 2.5th and 97.5th percentiles)
lower_bound = np.quantile(bootstrapped_means, 0.025)  # 2.5th percentile
upper_bound = np.quantile(bootstrapped_means, 0.975)  # 97.5th percentile

# Print the 95% confidence interval for the mean
print(f"95% Bootstrap Confidence Interval for the Mean: [{lower_bound}, {upper_bound}]")
```

```python
# Create a histogram of the bootstrapped means

import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))
plt.hist(bootstrapped_means, bins=30, edgecolor='black', alpha=0.7)
plt.title('Distribution of Bootstrapped Means')
plt.xlabel('Bootstrapped Mean')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)

# Show the mean and confidence interval lines
plt.axvline(np.mean(bootstrapped_means), color='red', linestyle='dashed', linewidth=1, label='Mean')
plt.axvline(lower_bound, color='blue', linestyle='dashed', linewidth=1, label='2.5% CI')
plt.axvline(upper_bound, color='blue', linestyle='dashed', linewidth=1, label='97.5% CI')
plt.legend()

plt.show()
```
